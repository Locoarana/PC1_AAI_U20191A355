```python
import matplotlib.pyplot as plt
import numpy as np
import random
import time
import math
```

```python
def read_file(filename="demo"):
  with open(filename) as f:
    n, m = [int(x) for x in f.readline().split()]
    jobs = np.zeros((n, m, 2), dtype=int)
    i = 0
    for line in f:
      jobs[i] = np.array([int(x) for x in line.split()]).reshape(m, 2)
      i += 1
  return n, m, jobs
```

```python
def all_jobs(rep):
  ttime = 0
  start, indices, end  = np.zeros((m), dtype=int), np.zeros((n), dtype=int), np.zeros((n), dtype=int)
  for e in rep:
    i = indices[e]
    indices[e] += 1
    midx = jobs[e, i, 0]
    time = jobs[e, i, 1]
    start[midx] = max(start[midx], end[e]) + time
    end[e] = start[midx]
    if start[midx] > ttime: ttime = start[midx]
  return ttime
```

```python
def random_sampling(minutes=3):
  rep = [i for i in list(range(n)) for _ in range(m)]
  t_end = time.time() + 60 * minutes
  while time.time() < t_end:
    makespan = all_jobs(random.sample(rep, len(rep))) #np.random.choice()
    random_sampling_iterations.append(makespan)
    random_sampling_schedules.append(rep)
```

```python
def graph():
  data = random_sampling()
  plt.title('Progress over time')
  plt.xlabel('Iterations')
  plt.ylabel('Makespans')
  plt.plot(random_sampling_iterations)
  plt.show()
```

```python
def gantt_graph():
  i = random_sampling_iterations.index(min(random_sampling_iterations))
  mstart = np.zeros((m), dtype=int)
  jend = np.zeros((n), dtype=int)
  idxs = np.zeros((n), dtype=int)
  schedule = np.zeros((m*10, min(random_sampling_iterations)))
  for e in random_sampling_schedules[i]:
      i = idxs[e]
      idxs[e] += 1
      midx = jobs[e, i, 0]
      time = jobs[e, i, 1]
      mstart[midx] = max(mstart[midx], jend[e])
      schedule[midx*10:(midx+1)*10, mstart[midx]:mstart[midx] + time] = e+1
      mstart[midx] += time
      jend[e] = mstart[midx]
  plt.figure(figsize=(10, 3))
  plt.imshow(schedule)
```

```python
class SimulatedA():
  def __init__(self, filename):
    self.n, self.m, self.jobs = read_file(filename)
    self.simulated_annealing_iterations = []
    self.random_sampling_schedules = []

  def all_jobs(self, rep):
    mstart = np.zeros((self.m), dtype=int)
    jend = np.zeros((self.n), dtype=int)
    idxs = np.zeros((self.n), dtype=int)
    makespan = 0
    for e in rep:
        i = idxs[e]
        idxs[e] += 1
        midx = self.jobs[e, i, 0]
        time = self.jobs[e, i, 1]
        mstart[midx] = max(mstart[midx], jend[e]) + time
        jend[e] = mstart[midx]
        if mstart[midx] > makespan:
            makespan = mstart[midx]
    return makespan
  
  def simulated_annealing(self, temperature, decrease, minutes=3):
    def neighbors(schedule):
      neighbors = []
      for i in range(len(schedule) - 1):
        neighbor = schedule[:]
        swapIndex = random.randrange(len(schedule))
        neighbor[i], neighbor[swapIndex] = neighbor[swapIndex], neighbor[i]
        neighbors.append(neighbor)
      return neighbors
    
    t_end = time.time() + 60 * minutes
    schedule = [i for i in list(range(self.n)) for _ in range(self.m)]
    random.shuffle(schedule)
    makespan = self.all_jobs(schedule)
    while time.time() < t_end:   
      temperature = decrease * float(temperature)
      for n in neighbors(schedule):
        new_makespan = self.all_jobs(n)
        if new_makespan < makespan:
          schedule = n
          makespan = new_makespan
          self.simulated_annealing_iterations.append((makespan))
          self.random_sampling_schedules.append(schedule)
        else:
          p = math.exp(-new_makespan/temperature)
          if random.random() < p:
            schedule = n
            makespan = new_makespan
            self.simulated_annealing_iterations.append((makespan))
            self.random_sampling_schedules.append(schedule)
  
  def graph(self):
    data = self.simulated_annealing(200, 0.8, .5)
    plt.title('Progress over time')
    plt.xlabel('Iterations')
    plt.ylabel('Makespans')
    plt.plot(self.simulated_annealing_iterations)
    plt.show()

  def gantt_graph(self):
    i = self.simulated_annealing_iterations.index(min(self.simulated_annealing_iterations))
    mstart = np.zeros((self.m), dtype=int)
    jend = np.zeros((self.n), dtype=int)
    idxs = np.zeros((self.n), dtype=int)
    schedule = np.zeros((self.m*10, min(self.simulated_annealing_iterations)))
    for e in self.random_sampling_schedules[i]:
        i = idxs[e]
        idxs[e] += 1
        midx = self.jobs[e, i, 0]
        time = self.jobs[e, i, 1]
        mstart[midx] = max(mstart[midx], jend[e])
        schedule[midx*10:(midx+1)*10, mstart[midx]:mstart[midx] + time] = e+1
        mstart[midx] += time
        jend[e] = mstart[midx]
    plt.figure(figsize=(10, 3))
    plt.imshow(schedule)
```
